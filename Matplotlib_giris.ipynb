# Giris

````
#import islemleri
import numpy as np 
import matplotlib.pyplot as plt

#bu, notebook icin kullanilmasi faydalidir buraya takilmaniza gerek yok
%matplotlib inline #bunu yazdıgımız için grafiği çizdirmek için plt.show() fonksiyonunu kullanmamıza gerek kalmıyor 
````

**Örnek**

````
#verileri liste seklinde  alalim
yas = [10,20,30,35,36,37,46,48,65,80 ]
kilo = [35,49,50,52,54,55,60,68,78,92]
````

Genel olarak numpy ile kullanim alani cok oldugu  ve baska islemlere daha acik oldugu icin listeleri numpy arraye cevirdik .Elimizdeki 2 listeyide numpy arraylerine cevirelim

````
numpy_yas = np.array(yas)
numpy_kilo = np.array(kilo)
````

Kontrol edelim

````
numpy_kilo
````

````
numpy_yas
````

Ilk cizdirmemizi yapalim <p>
plt.plot´a ilk olarak x degerlerini ,ikinci parametre olarak da y degerlerini gonderelim <p>
tirnak icnde yazdigimiz ise grafigin rengiyle alakalidir

````
plt.plot(numpy_yas,numpy_kilo,"r")
plt.xlabel("Yas") #X EKSENI ISMI
plt.ylabel("Kilo") #y ekseni ismi
plt.title("Kilo Vs Yas")# genel baslik
plt.show() #bunu notebookta yazmak zorunda degiliz ama normal programlarimizda bunu yazariz gostermek icin
````
